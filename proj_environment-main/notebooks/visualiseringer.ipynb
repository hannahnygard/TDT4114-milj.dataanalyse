```python
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Lufttrykk:
-En visualisering av lufttrykk, med illustrasjoner av median, gjennomsnitt og standardavvik

```python
#henter ut de rensede verdiene fra pickle filen
with open("../data/renset_data_lufttrykk.pkl", "rb") as f:
    df_lufttrykk = pickle.load(f)


#Konvertere datoene fra dataframen til dato-format 
df_lufttrykk["date"] = pd.to_datetime(df_lufttrykk["date"])

pd.set_option('display.max_rows', 1000)  # Øker antall rader som vises
pd.set_option('display.max_columns', 20)  # Øker antall kolonner som vises
pd.set_option('display.width', 1000)  # Øker bredden på utskriften

#Selve visualiseringen 
sns.scatterplot(x="date", y="air_pressure", data=df_lufttrykk)

#Legger inn gjennomsnittet
plt.axhline(y=1010, color="r")
```

Tabell Nedbør renset:
- Fjerner unødvendig informasjon. Freks: klokkeslett, sourceID, timeOffset etc..
- Regner ut gjennomsnittet av begge nedbørsmålingene som skjer på samme dag, slik at det blir ryddigrere og mer oversiktelig.
- Satt inn dato 31.12.2023, og gitt den medianverdien til desember 2023


```python
def fra_json_til_dataframe(filnavn):
    with open(filnavn, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    #Henter inn observasjonene
    observationer = data.get("data", [])
    
    rader = []

    for obs in observationer:
        source_id = obs.get('sourceId', None)
        reference_time = obs.get('referenceTime', None)

        #For hver observasjon vi gjør lager vi en rad med nødvendig informasjon
        for observation in obs.get('observations', []):
            rader.append({
                'sourceId': source_id,
                'referenceTime': reference_time[:10], #Tar bort klokkeslett, har bare med dato
                'elementId': observation.get('elementId', None),
                'value': observation.get('value', None),
                'unit': observation.get('unit', None)
            })
    
    df = pd.DataFrame(rader)

    #Trenger at value blir numerisk, slik at vi kan regne ut gjennomsnittet
    df["value"] = pd.to_numeric(df["value"], errors="coerce")
    return df

df_nedbor = fra_json_til_dataframe("frost_nedbor.json")

#Leser dataene etter dato, før vi deretter beregner gjennomsnittet
df_nedbor_avg = df_nedbor.groupby("referenceTime", as_index=False)["value"].mean()
df_nedbor_avg["unit"] = df_nedbor.groupby("referenceTime")["unit"].first().reset_index(drop=True)

#Telle duplikater før de fjernes
antall_duplikater = df_nedbor_avg.duplicated().sum()
print(f"Antall duplikater funnet: {antall_duplikater}")

#Dersom dataen har duplikater, blir de fjernet
df_nedbor_avg.drop_duplicates(inplace=True)

#Sjekker om det mangler noen datoer
datoer_2023 = pd.date_range(start="2023-01-01", end="2023-12-31").strftime("%Y-%m-%d")
manglende_datoer = set(datoer_2023) - set(df_nedbor_avg["referenceTime"])

#Regner ut median
def median():
    desember_data = df_nedbor_avg[df_nedbor_avg["referenceTime"].astype(str).str.startswith("2023-12")]
    median_desember = desember_data["value"].median().round(2)
    return median_desember
 
median_desember = median()
print("Median for temperaturen i desember er: ", median_desember)

#legger til ny rad for den manglende datoen
ny_rad = pd.DataFrame({
    "referenceTime": sorted(list(manglende_datoer)),
    "value": [median_desember] * len(manglende_datoer),
    "unit": [df_nedbor_avg["unit"].mode()[0]] * len(manglende_datoer)
})

df_nedbor_avg = pd.concat([df_nedbor_avg, ny_rad], ignore_index=True)

#Skjekker om det er noen datoer som mangler en verdi
manglende_nedbor = df_nedbor_avg["value"].isna().sum()

if manglende_nedbor > 0:
    print(f"Antall datoer med manglende verdi: {manglende_nedbor}")
    print("Datoer uten verdi:")
    print(df_nedbor_avg[df_nedbor_avg["value"].isna()])
else:
    print("Det er ingen datoer som mangler verdier!")

#Utskiften
print("Gjennomsnittlig nedbør per dag i 2023:")
display(df_nedbor_avg)
```

Tabell temperatur renset:
- Fjerner unødvendig informasjon. Feks: klokkeslett, sourceID, timeOffset etc..
- Fjerner 2 desimaler, slik at det bare er 1 som skrives ut
- Ser og fjerner eventuelle duplikater
- Skjekker at alle år har en verdi

```python
def fra_json_til_dataframe(filnavn):
    with open(filnavn, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    # Henter observasjonene
    observationer = data.get("data", [])
    
    rader = []

    for obs in observationer:
        reference_time = obs.get('referenceTime', None)

        #For hver observasjon lager vi en rad med nødvendig informasjon
        for observation in obs.get('observations', []):
            rader.append({
                'referenceTime': reference_time[:4],  #Tar bare med årstall
                'value': observation.get('value', None),
                'unit': observation.get('unit', None)
            })
    
    df = pd.DataFrame(rader)
    return df

df_temp = fra_json_til_dataframe("frost_temp.json")

#Runder av til 1. desimal
df_temp["value"] = df_temp["value"].round(1)

#Endrer navn på kolonne "refrenceTime" til year
df_temp.rename(columns={"referenceTime" : "year"}, inplace=True)

#Legger til unit i utskrifte
unit = df_temp["unit"].dropna().unique()
df_temp["unit"] = unit[0] if len(unit) > 0 else "Ukjent"

#Telle duplikater før de fjernes
antall_duplikater = df_temp.duplicated().sum()
print(f"Antall duplikater funnet: {antall_duplikater}")

#Dersom dataen har duplikater, blir de fjernet
df_temp.drop_duplicates(inplace=True)

#Skjekker om det er noen datoer som mangler en verdi
manglende_temperaturer = df_temp["value"].isna().sum()

if manglende_temperaturer > 0:
    print(f"Antall datoer med manglende verdi: {manglende_temperaturer}")
    print("Datoer uten verdi:")
    print(df_temp[df_temp["value"].isna()])
else:
    print("Det er ingen år som mangler verdier!")

#Utskrift
print("Gjennomsnittlig temperatur per år fra 1980-2020:")
display(df_temp)
```

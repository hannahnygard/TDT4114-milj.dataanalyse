```python
import pandas as pd
import sqlite3

#noe kode inspirert av w3-schools, men ikke direkte kopiert

#Leser dataen fra json filen
df = pd.read_json("frost_skydekke.json")

#Konverter dataen til en dict også videre til en todimensjonal tabell
df = pd.json_normalize(df.to_dict(orient="records"))

#Går gjennom kolonne for kolonne. Dersom x er en liste eller dict blir den konvertert til en streng, er den ikke det forblir x uendret
df = df.apply(lambda kolonne: kolonne.apply(lambda x: str(x) if isinstance(x, (dict, list)) else x))

#Bruker modulen sqlite3 for å koble oss opp til databasen vår
kobling = sqlite3.connect("frost_Database.db")

#Lagrer dataframen som en sql tabell
df.to_sql("tabell", kobling, if_exists= "replace", index=False)

kobling.close()

print("Suksess! JSON er nå omgjort til en database")






```

```python
#Fikk ikke ut alle radene i tabellen ovenfor (data inni dataobservations gjemmer seg inni en ny dict), disse ønsker vi å få med i printet
#Kode inspirert av pandas dokumentasjon (pandas.pydata.org)
kobling = sqlite3.connect("frost_Database.db")

cursor = kobling.cursor() #lager et cursor objekt for å utføre sql spørringer
cursor.execute("SELECT * FROM tabell LIMIT 5")      # henter kun ut de første 5 radene
rows = cursor.fetchall()     #henter ut alle radene
for row in rows:        #itererer gjennom en og en rad og printer dette ut
    print(row)

kobling.close()
```

```python
import sqlite3
import pandas as pd
import ast  #bruker ast for å konvertere strenger til lister
 
# Kobler til databasen
kobling = sqlite3.connect("frost_Database.db")
 
# Hent ut tidspunkt og observasjoner fra tabellen
df = pd.read_sql_query("SELECT `data.referenceTime`, `data.observations` FROM tabell", kobling)
 
# Lukker koblingen
kobling.close()
 
# Konverter 'data.observations' fra streng til liste
df["data.observations"] = df["data.observations"].apply(ast.literal_eval)
 
# Henter ut verdien av lufttrykket og målenhet fra observations-listen
df["values"] = df["data.observations"].apply(lambda x: x[0]["value"] if isinstance(x, list) and len(x) > 0 else None)
df["unit"] = df["data.observations"].apply(lambda x: x[0]["unit"] if isinstance(x, list) and len(x) > 0 else None)


# Behold ku, sau og lufttrykk
df = df[["data.referenceTime", "values", "unit"]]
df = df.rename(columns={"data.referenceTime": "year"})

 
# Skriv ut resultatet
display(df)

 
```

**Rensing av data**
- Fjerning av duplikater
- Formatering av dato (er kun ute etter en dag, ikke klokkeslett)
- Sjekker etter manglende datoer

```python
#Endrer rekkefølgen på kolonnene
df = df[["year", "values", "unit"]]

df["year"] = pd.to_datetime(df["year"], utc=True).dt.year


#sjekker om det er noen datoer som mangler
årstall = set(range(1980, 2021))
manglende_år = årstall - set(df["year"])

#Itererer gjennom de manglende datoene og skriver dem ut
if manglende_år:
    print("Manglende datoer i datasettet:")          
    for år in sorted(manglende_år):
        print(år)
else:
    print("Ingen datoer mangler:-)")                  


#Printer ut duplikater, dersom dette finnes - kode hentet fra pensum-notebooken
print("Antall duplikater:", df.duplicated().sum())

#Sjekker om det er noen datoer som mangler en verdi - kode hentet fra pensum-notebooken
manglende_data = df["values"].isna().sum()
if manglende_data > 0:
    print(f"Antall datoer med manglende verdi: {manglende_data}")
    print("Datoer uten verdi:")
    print(df[df["values"].isna()])
else:
    print("Ingen av datoene har manglende verdier!")


#Skriver ut resultatet
display(df)

```
